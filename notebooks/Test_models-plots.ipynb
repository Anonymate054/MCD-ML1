```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Datos de la tabla
data = {
    'Starting': [100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800, 900, 950],
    'Ending': [150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800, 850, 950, 1000],
    'QR_loss_mean_50': [8.23913, 5.75086, 9.04918, 13.73446, 17.27828, 14.72398, 25.51823, 27.85007, 27.03975, 
                      30.57936, 31.88254, 80.7059, 37.35936, 59.73315, 37.95264, 79.19228, 89.51612],
    'QR_loss_mean_75': [6.80128, 5.03588, 8.60884, 12.75963, 14.03208, 17.09278, 24.99835, 26.03692, 25.12579, 
                      34.29613, 31.79135, 124.27891, 41.57112, 95.30279, 150.67322, 115.39633, 126.44683],
    'Catboost_RMSE': [ 16.58511,  14.80152,  19.65784,  20.62923,  23.63782,  22.32577,  35.77981,
  53.77349,  56.03752,  41.1777,   65.27986, 135.50864,  88.17994, 141.58072, 174.38472, 139.95672, 276.31201],
    'Count': [9, 62, 93, 97, 69, 36, 39, 17, 13, 10, 4, 5, 2, 7, 2, 2, 1]
}

# Creaci√≥n del DataFrame
df = pd.DataFrame(data)
```

```python
# Crear gr√°fico de barras y l√≠neas
fig, ax1 = plt.subplots(figsize=(14, 6))

# Gr√°fico de barras para el Count
ax1.bar(df['Starting'], df['Count'], width=40, color='lightgrey', label='Count of Cars')

# Creaci√≥n de segundo eje Y para los errores
ax2 = ax1.twinx()

# Gr√°fica de l√≠neas para QR_loss_mean_50
ax2.plot(df['Starting'], df['QR_loss_mean_50'], color='blue', label='Pinball Loss 50', marker='o')

# Gr√°fica de l√≠neas para QR_loss_mean_75
ax2.plot(df['Starting'], df['QR_loss_mean_75'], color='red', label='Pinball Loss 75', marker='o')

# Gr√°fica de l√≠neas para Catboost
ax2.plot(df['Starting'], df['Catboost_RMSE'], color='green', label='RMSE', marker='o')

# T√≠tulo y etiquetas
ax1.set_xlabel('Price Range Starting (in $k)')
ax1.set_ylabel('Count of Cars')
ax2.set_ylabel('Pinball Loss & RMSE (in $k)')
# plt.title('Comparison of Car Distribution and Prediction Errors (QR Loss) Across Price Ranges')
plt.title('Comparison of Car Distribution and Prediction Errors Across Price Ranges')


# Leyendas
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Mostrar gr√°fico
plt.show()
```
